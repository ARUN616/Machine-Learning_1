```python
import pandas as pd
```

```python
data=pd.read_csv('insurance_pre.csv')
```

```python
data
```

```python
data=pd.get_dummies(data,drop_first=True)
```

```python
data
```

```python
data.columns
```

```python
independent=data[['age', 'bmi', 'children', 'sex_male', 'smoker_yes']]
dependent=data[['charges']]
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(independent,dependent,test_size=1/3,random_state=0)
```

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
X_train=sc.fit_transform(X_train)
X_test=sc.transform(X_test)
```

```python
from sklearn.svm import SVR
from sklearn.model_selection import GridSearchCV
```

```python
param_grid={'kernel':['linear', 'poly', 'rbf', 'sigmoid'],'gamma':['scale','auto'],'C':[1,10,100,1000,10000]}
grid = GridSearchCV(SVR(), param_grid, refit = True, verbose = 3,n_jobs=-1) 
grid.fit(X_train,y_train)
```

```python
re=grid.cv_results_
re
```

```python
table=pd.DataFrame.from_dict(re)
table
```

```python
grid_predictions=grid.predict(X_test)
from sklearn.metrics import r2_score
r_score=r2_score(y_test,grid_predictions)
print(f'r2_score for the best parameter of SVMR is : ',grid.best_params_,' and r_score is: ',r_score)
```
